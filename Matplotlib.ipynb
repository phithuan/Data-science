### pyplot API

```python
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x), color="blue", linestyle="dashed", label="sin(x)");
plt.plot(x, np.cos(x), color="red", label="cos(x)")

plt.title("Sin(X) and Cos(X)")
plt.xlabel("X");
plt.ylabel("F(x)");

#plt.xlim([0,4]) #giới hạn trục x từ 0 - 4
#plt.ylim([-0.6, 0.6]) # giới hạn trục y từ -1 - 1
#plt.axis([0,4, -0.6, 0.6])         CÁCH 2
#plt.axis("tight")
plt.axis("equal")


plt.legend();# gọi lên này ra mới hiện label
#plt.show(); when run on terminal
```

### Object-Oriented API

```python
%matplotlib inline

#1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#2. Setup plot
fig, ax = plt.subplots(figsize=(5,5)) #Figure size = Width & Height of the plot
# fig là hình tổng, ax là khung hình nhỏ
#3. Plot the data
ax.plot(x, y);

#4. Customize the plot
ax.set(title="A simple plot",
       xlabel="x-axis",
       ylabel="y-axis");
```

# Project: Data Visualization - California Cities

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("california_cities.csv")
df.head()

#df["area_total_km2"].sort_values()
```

```python
# Extract Latd = Vĩ Độ and Longd = Kinh Độ
lat, lon = df['latd'], df['longd']
population, area = df['population_total'], df['area_total_km2']

plt.figure(figsize=(8,6))

# Plot using Pyplot API
plt.scatter(lon, lat, 
            c=np.log10(population), cmap='viridis',
            s=area, linewidths=0, alpha=0.5)

plt.title('California Cities: Population and Area Distribution');
plt.axis('equal')
plt.xlabel('longlatitude')
plt.ylabel('latitude');
plt.colorbar(label='log10(population)')
plt.clim(3,7)


#Create a legend for cities'sizes
area_range = [50, 100, 300, 500]

for area in area_range:
    plt.scatter([], [],  s=area, c='k', alpha=0.5,
                label=str(area) + ' km$^2$')

plt.legend(labelspacing=1, title='City Area')
```
